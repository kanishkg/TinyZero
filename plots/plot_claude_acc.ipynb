```python
# Load the data
df = pd.read_csv("../results/claude_accs.csv")
```

```python
import matplotlib.patches as mpatches

metrics = df['condition'].tolist()
accuracies = df['average_score'].tolist()
x = np.arange(len(metrics))
width = 0.25  # Width of each bar

fig, ax = plt.subplots(figsize=(15, 10))

rects1 = ax.bar(x - width, accuracies, width, label='Qwen Base', alpha=0.9, facecolor=deep[1])


# Customize plot
ax.set_xticks([])
ax.spines["left"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.yaxis.grid(False)
ax.set_ylim(0, 1.)
ax.grid(axis='y', color='gray', alpha=0.2, linestyle='-')
ax.tick_params(axis='y', which='both', length=0)

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)


plt.tight_layout()
# save
plt.savefig('claude_accuracies.png', dpi=300, bbox_inches='tight')
```
